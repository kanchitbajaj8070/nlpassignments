```python
def preprocessing_data( paragraph):
    paragraph=paragraph.lower()
    corpus=[]
    sentences= sent_tokenize(paragraph)
    for s in sentences:
        words=reg.tokenize(s)
        words=[ lemmatizer.lemmatize(w) for w in words if w not in set(stopwords.words('english'))]
        corpus.append(' '.join(words))
    return corpus
        
```

```python
corpus=preprocessing_data(paragraph)

corpus
```

```python
def preprocessing_data_word2vec( paragraph):
    paragraph=paragraph.lower()
    corpus=[]
    sentences= sent_tokenize(paragraph)
    for s in sentences:
        words=reg.tokenize(s)
        words=[ w for w in words if w not in set(stopwords.words('english'))]
        corpus.append(words)
    return corpus
        
```

```python
corpus=preprocessing_data_word2vec(paragraph)
corpus
```

```python
model=Word2Vec(corpus,min_count=1)
```
