```python
def vocab_size(text)->int:
    unique_words=set([w for w in text.split(" ") if w not in [',','.']])
    return len(unique_words)
```

```python
print(vocab_size("my name is kunal bajaj. i am in thapar university . i like the nlp course"))
```

```python
def percent(word, text)-> int :
    freq=len([w for w in text if w==word])
    total_words=vocab_size(text)
    return (freq/total_words)*100
```
